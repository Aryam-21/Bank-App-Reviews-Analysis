# Bank Review Scraping & Preprocessing
This notebook runs Taskâ€‘1: Web Scraping and Preprocessing Pipeline.

```python
import sys
import os
project_root = os.getcwd()  
sys.path.append(project_root)
import pandas as pd
from scraper import PlayStoreScraper
from processing import ReviewPreprocessor
print(sys.path)
```

## Scrape Reviews

```python
scraper = PlayStoreScraper()
all_reviews = []

for bank, app_id in scraper.app_ids.items():
    print(f"\nScraping: {bank}")
    data = scraper.scrape_reviews_all(app_id)
    processed = scraper.process_reviews(data, bank)
    all_reviews.extend(processed)
data = pd.DataFrame(all_reviews)
data.to_csv('../data/raw/reviews_raw.csv', index=False)
data.head()
```

## Preprocess Data

```python
INPUT_PATH = '../data/raw/reviews_raw.csv'
OUTPUT_PATH = '../data/processed/reviews_processed.csv'
processor = ReviewPreprocessor(input_path=INPUT_PATH, output_path=OUTPUT_PATH)
processor.run()
pd.read_csv('../data/processed/reviews_processed.csv').head()
```
